```python
import json
from pprint import pprint 
from pymongo import MongoClient
from pymongo.errors import *

client = MongoClient('localhost',27017)
db = client['books']
info_books = db.info_books

for book in info_books.find({"_id": 0}):
    pprint(book)
```

```python
for book in info_books.find({'price':{'$lt': 10.1}}):
    pprint(book)
```

```python
count = info_books.count_documents({'price': {'$gt': 10.1}})
print(count)
```

```python
for book in info_books.find({'price': {'$gte': 5, '$lte': 10}}):
    pprint(book)
```

```python
for book in info_books.find({'availability': 'In stock', 'price': {'$lt': 10.1}}):
    pprint(book)
```

```python
for book in info_books.find().sort('price'):
    pprint(book)
```

```python
for book in info_books.find().sort('price', -1):
    pprint(book)
```

```python
for book in info_books.find({"_id": 0}):
    pprint(book)
    
info_books.update_one({'_id': 0}, {'$set': {'price': 12.99}})

for book in info_books.find({"_id": 0}):
    pprint(book)
```
